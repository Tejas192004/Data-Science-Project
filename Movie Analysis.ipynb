```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("IMDb_Dataset_2.csv")
data
```

```python
data.dropna()
```

```python
data.fillna(data.mean())
```

```python
long_movies=data[data['Duration (minutes)']>=180]
print(long_movies['Title'])
```

```python
sns.countplot(x=data['Genre'])
plt.xlabel("Types of movies")
plt.xticks(data['Genre'])

plt.show()
```

```python
#This graph shows that Biography movies are created most

```

```python
sns.countplot(x=data['IMDb Rating'])
plt.xlabel("Types of Rating")
plt.xticks(rotation=720)
plt.show()
```

```python
average_votes = data.groupby('Year')['IMDb Rating'].max()
average_votes
```

```python
movies_per_year = data.groupby('Year')['Title'].count()
print(movies_per_year)
```

```python
sns.barplot(x=data['Title'],y=data['Duration (minutes)'],data=long_movies)
plt.show()
```

```python
sns.countplot(x=data['Director'])
plt.xticks(rotation=90)
plt.show()
```

```python
average_rating_per_year=data.groupby('Year')['IMDb Rating'].mean()
print(average_rating_per_year)
```

```python
top_10_rated_movies=data.sort_values(by='IMDb Rating',ascending=False).head(10)
print(top_10_rated_movies)
```

```python
data.drop(['Poster-src'],axis=1)
```

```python
# Visualize data
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(data['Duration (minutes)'], bins=50)
plt.title('Duration Distribution')
plt.xlabel('Duration (minutes)')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
plt.scatter(data['Duration (minutes)'], data['IMDb Rating'])
plt.title('Duration vs. Rating')
plt.xlabel('Duration (minutes)')
plt.ylabel('Average Rating')

plt.tight_layout()
plt.show()
```
